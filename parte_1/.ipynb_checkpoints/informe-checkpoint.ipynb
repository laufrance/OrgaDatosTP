## ¿A mayor humeadad, más probabilidades de que lluevan hamburguesas al día siguiente?

A priori, uno podria especular que los features que son más relevantes a la lluvia de hamburguesas son los que tienen relación con la lluvia normal, es decir, la humedad, los milímetros de lluvia al día, los milimetros evaporados de agua, entre otros. ¿Será que también contribuyen a la lluvia de hamburguesas? Comenzemos a analizar algunos de estos features, empezando por la humedad.

```python
humedades = datos_lluvia.loc[:,['humedad_tarde', 'humedad_temprano']]

plt.figure(dpi=120)

plt.hist(
    humedades['humedad_tarde'],
    bins=50,
    label="Humedad a la tarde",
    alpha=0.5,
    color ='#F1B388',
)
plt.hist(
    humedades['humedad_temprano'],
    bins=50,
    label="Humedad temprano",
    alpha=0.5,
    color='#C6F188',
)

plt.title("Distribución de la humedad a la tarde y temprano")
plt.xlabel("% de humedad")
plt.ylabel("Cantidad")
plt.legend()

plt.show()
```

Lo llamativo es que vemos un mayor porcentaje de humedad temprano, con un pico en 100%. Luego veremos si esto es relevante o no. Veamos ahora sí la relacion entre las humedades a la mañana y a la tarde con si llovieron hamburguesas al otro día.

## ¿A menos horas de sol, más lluvia al día siguiente?

Antes que nada, este feature tiene muchísimos NaNs, casi la mitad:

```python
datos_lluvia['horas_de_sol'].isna().value_counts()
```

¿Qué tal si estos missings se relacionan con la lluvia de hamburguesas al día siguiente? En ese caso, si los omitimos nos estaríamos perdiendo información muy valiosa... veamos si esconden algún patrón.

```python
horas_sol_missings = datos_lluvia[['llovieron_hamburguesas_al_dia_siguiente', 'horas_de_sol']]
horas_sol_missings = horas_sol_missings[horas_sol_missings['horas_de_sol'].isna()]
horas_sol_missings['horas_de_sol'] = horas_sol_missings[['horas_de_sol']].fillna('missing')

plt.figure(dpi=130)
horas_sol_missings['llovieron_hamburguesas_al_dia_siguiente'].value_counts().sort_index().plot(kind='pie', autopct='%1.0f%%', colors=['#F1B388', "#C6F188"])
plt.title('Si no hay info de las horas de sol, ¿llueven hamb. al día siguiente?')
plt.ylabel("")
plt.show()
```

Parece que no hay relación, no tiene nada que ver si no hay información de las horas de sol con si llovieron hamburguesas al día siguiente porque los porcentajes son muy similares a los que vimos en la introducción.

Pero ahora veamos si los valores no-null de las horas de sol se relacionan con si llovieron hamburguesas al día siguiente.

```python
horas_y_lluvias = datos_lluvia.loc[:,['horas_de_sol', 'llovieron_hamburguesas_al_dia_siguiente']]

plt.figure(dpi=130)
plt.title("Relacion entre las horas de sol y si llovio al dia siguiente")
sns.violinplot(
    data=horas_y_lluvias,
    y='horas_de_sol',
    x='llovieron_hamburguesas_al_dia_siguiente',
    palette=['#F1B388', "#C6F188"],
)
plt.ylabel("Horas de Sol")
plt.xlabel("Llovieron hamburguesas al día siguiente")
plt.show()
```

Este grafico nos deja muy en claro que si el día tuvo muchas horas de sol, hay una alta probabilidad de que no llovieron hambuguresas al día siguiente, pero parece que si hay pocas horas de sol **no** podemos decir con certeza que va a llover. Veámoslo de otra forma:

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=[6.4 * 2, 4.8], dpi=100, sharey=True)

datos_llovio = datos_lluvia[datos_lluvia['llovieron_hamburguesas_al_dia_siguiente'] == 'si']
datos_no_llovio = datos_lluvia[datos_lluvia['llovieron_hamburguesas_al_dia_siguiente'] == 'no']

datos_llovio['horas_de_sol'].plot(
    kind='hist', ax=axes[0], xlim=datos_lluvia.horas_de_sol.min(), bins=50, color= "#F1B388"
)
axes[0].set_title("Horas de Sol - Sí llovío al día siguiente")
axes[0].set_ylabel("Cantidad")
axes[0].set_xlabel("Horas de sol")

datos_no_llovio['horas_de_sol'].plot(
    kind='hist', ax=axes[1], xlim=datos_lluvia.horas_de_sol.min(), bins=50, color= '#C6F188'
)
axes[1].set_title("Horas de Sol - No llovío al día siguiente")
axes[1].set_ylabel("Cantidad")
axes[1].set_xlabel("Horas de sol")

plt.show()
```

Confirmamos lo que vimos antes. Podríamos decir que si hubieron más de 9 horas de sol, casi seguro no llovió al día siguiente. También, si hubo menos de 6 horas de sol, el gráfico verde acumula relativamente poca probabilidad a izquierda, mientras que el naranja acumula la mayoría de la probabilidad. Así que estos dos parámetros van a ser buenos para el baseline.

## ¿Cómo se relaciona la lluvia o no de hamburguesas de hoy con la de mañana?

Nos intriga esta relación porque a veces la lluvia se extiende a lo largo de varios días, así como también los períodos donde no llueve. ¿Será que si hoy llovió, las chances de que llueva mañana son altas también? ¿Si hoy no llovió, con cuánta certeza podemos decir que mañana tampoco lloverá?

```python
lluvias = datos_lluvia[['llovieron_hamburguesas_al_dia_siguiente', 'llovieron_hamburguesas_hoy']]
llovio_hoy = lluvias.loc[lluvias['llovieron_hamburguesas_hoy'] == 'si']
no_llovio_hoy = lluvias.loc[lluvias['llovieron_hamburguesas_hoy'] == 'no']

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=[6.4 * 2, 4.8], dpi=100)

no_llovio_hoy['llovieron_hamburguesas_al_dia_siguiente'].value_counts()[
    ["no", "si"]
].sort_index().plot(kind='pie', autopct='%1.0f%%', colors=['#F1B388', "#C6F188"], ax=axes[0])
axes[0].set_title("Si hoy no llovió, entonces ¿mañana llueve?")

llovio_hoy['llovieron_hamburguesas_al_dia_siguiente'].value_counts()[
    ["no", "si"]
].sort_index().plot(kind='pie', autopct='%1.0f%%', colors=['#F1B388', "#C6F188"], ax=axes[1])
axes[1].set_title("Si hoy sí llovió, entonces ¿mañana llueve?")
                    
plt.ylabel("")
plt.show()
```

Si hoy llovió, no podemos tener un indicio claro de qué pasó al día siguiente, pues el gráfico de la derecha muestra un 50/50 casi. Pero el gráfico de la izquierda practicamente nos deja asumir que si hoy no llovió, mañana no va a llover -y eso va a estar bien en un 85% de las veces. 

## ¿Habrá alguna relación entre los mm_de_lluvia_dia y la lluvia de hamburguesas?

Queremos ver si a partir de ciertos milimetros podemos decir que hay alta probabilidad que lluevan hamburguesas al día siguiente. Hagamos un describe para tener una idea de los datos de este feature.

```python
mm_y_hamburguesas_al_dia_siguiente = datos_lluvia.loc[:,['mm_lluvia_dia', 'llovieron_hamburguesas_al_dia_siguiente']]
mm_y_hamburguesas_al_dia_siguiente.describe()
```

La mediana es 0, hasta el 3° cuartil es casi 0. Veamos como es la distribución. Es lógico porque en general no llueve.

```python
mm_y_hamburguesas_al_dia_siguiente = mm_y_hamburguesas_al_dia_siguiente.loc[mm_y_hamburguesas_al_dia_siguiente['mm_lluvia_dia'] < 12]
```

Acotamos el espacio de representacion ya que hay unos pocos valores que son exagerados y destacan de los demas, lo que hacen dificil poder plotear de forma adecuada.

```python
plt.figure(dpi=130)
plt.title("Relación entre los mm de lluvia al día y si llovieron hamburguesas al día siguiente")
sns.boxplot(
    data = mm_y_hamburguesas_al_dia_siguiente,
    y='mm_lluvia_dia',
    x='llovieron_hamburguesas_al_dia_siguiente',
    palette=['#F1B388', "#C6F188"],
)
plt.ylabel("mm_lluvia_dia")
plt.xticks([False, True], ["No", "Sí"])
plt.show()
```

Se confirmó lo que sospechábamos. Ahora, intentemos analizar la información y cómo podríamos usar este feature en el baseline, a través de la probabilidad de que llueva o no al día siguiente dado que al día anterior hubo más o menos de una cierta cantidad de mm de agua llovidos.

```python
def plot_pie_mm_lluvia(ax, mm, greaterThan=True):
    if greaterThan:
        datos_mm_lluvia = datos_lluvia[datos_lluvia['mm_lluvia_dia'] > mm]
    else:
        datos_mm_lluvia = datos_lluvia[datos_lluvia['mm_lluvia_dia'] < mm]
    
    datos_mm_lluvia['llovieron_hamburguesas_al_dia_siguiente'].value_counts().sort_index().plot(
    kind='pie', ax=axes[ax], ylabel="", autopct='%1.0f%%', colors=['#F1B388', "#C6F188"])
    axes[ax].set_title(f'... {"> " if greaterThan else "< "} {mm}? (Total: {len(datos_mm_lluvia)})')

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=[6.4 * 2, 4.8], dpi=100)

plot_pie_mm_lluvia(ax=0, greaterThan=False, mm=0.7)

plot_pie_mm_lluvia(ax=1, mm=0.7)

plot_pie_mm_lluvia(ax=2, mm=50)

plot_pie_mm_lluvia(ax=3, mm=150)

fig.suptitle('¿Llovieron hamburguesas al día siguiente cuando los mm llovidos del día anterior eran...', fontsize=16)
plt.show()
```

Acá se puede apreciar que si llovió menos de 1 mm, es muy alta la probabilidad de que al día siguiente no hayan llovido hamburguesas. Esto es clave para el baseline. Por otro lado, si llovió más de 50 mm aumenta mucho la probabilidad de que sí llueva, pero el problema ahí es que son poquitas las instancias que eventualmente clasificaríamos correctamente en el baseline.

¿Será que a mayor nubosidad, mayor chance de que lluevan hamburguesas al día siguiente?

```python
datos_lluvia.nubosidad_tarde.value_counts()
```

```python
datos_lluvia.nubosidad_temprano.value_counts()
```

## Presiones atmosfericas, ¿influyen?

Hagamos un gráfico de dispersión para ver si existe algún indicio de que pueda haber información útil...

```python
plt.subplots(dpi=130)

sns.scatterplot(
        x=datos_lluvia['presion_atmosferica_tarde'],
        y=datos_lluvia['presion_atmosferica_temprano'],
        s=10,
        alpha = 1,
        hue = datos_lluvia['llovieron_hamburguesas_al_dia_siguiente'],
        palette=['#F1B388', "#C6F188"] 
    )

plt.legend(title="Llovió al día siguiente?")
plt.ylabel("Presión atmosferica a la mañana")
plt.xlabel("Presión atmosferica a la tarde")
plt.title('Relación entre la presion atmosferica y si llovieron hamburguesas al día siguiente')
plt.show()
```

Aca podemos apreciar que si las presiones atmosfericas a la mañana y a la tarde se mantienen > 1010, entonces podemos tener un indicio de que no llueve al día siguiente. En cambio, si las presiones están por debajo de 1010, las chances de que lluevan son mucho mayores...

Veamos estos datos dispersados en graficos de torta, para poder analizarlos mejor

```python
def plot_pie_presion(ax, presion, greaterThan=True):
    if greaterThan:
        datos_presion_lluvia = datos_lluvia[datos_lluvia['presion_atmosferica_tarde'] > presion]
    else:
        datos_presion_lluvia = datos_lluvia[datos_lluvia['presion_atmosferica_tarde'] < presion]
    
    datos_presion_lluvia['llovieron_hamburguesas_al_dia_siguiente'].value_counts().sort_index().plot(
    kind='pie', ax=axes[ax], ylabel="", autopct='%1.0f%%', colors=['#F1B388', "#C6F188"])
    axes[ax].set_title(f'... {"> " if greaterThan else "< "} {presion}? (Total: {len(datos_presion_lluvia)})')

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=[6.4 * 2, 4.8], dpi=100)

plot_pie_presion(ax=0, greaterThan=False, presion=1005)

plot_pie_presion(ax=1, presion=1010)

plot_pie_presion(ax=2, presion=1017)

plot_pie_presion(ax=3, greaterThan=False, presion=1000)

fig.suptitle('¿Llovieron hamburguesas al día siguiente cuando la presión atmosférica a la tarde era...', fontsize=16)
plt.show()
```

Se confirmó lo que dijimos antes. Además, sacamos algunos datos que nos van a servir en el baseline. En conclusión, sí, podemos afirmar que hay una relación bastante importante entre la presión atmosférica y la lluvia al día siguiente.

## ¿Cómo influye la temperatura en si llovieron hamburguesas al día siguiente?

Este grafico no nos deja un claro espacio en el que la temperatura marque si llueve o no. La temperatura parece no tener mucha influencia en si llueve o no al dia siguiente. # Función Baseline

Analizando todos estos resultados podemos armar una función BASELINE la cual nos devuelva una lista que tenga un accuracy elevado para poder predecir las lluvias de hamburguesas provocada por la máquina de Flint al día siguiente de una determinada instancia.

Básicamente, pensamos en esta lógica: de entrada, clasificamos directamente algunas instancias como 0 ó como 1 por parámetros fuertes que vimos en el análisis. Luego, hicimos un sistema de acumulación de condiciones en las que si se daban como mínimo una cantidad de circunstancias, era catalogado como 1. El resto de los valores, eran 0, ya que la mayoría del dataset es 0. Veamos como queda:

```python
def baseline(dataframe):
    guess = []
    for index, row in dataframe.iterrows():
        if (row['llovieron_hamburguesas_al_dia_siguiente'] == "si") | (row['llovieron_hamburguesas_al_dia_siguiente'] == "no"): 
            count = 0
            if row['horas_de_sol'] >= 7 and row['humedad_tarde'] <= 70: #Descartamos de entrada algunas instancias
                guess.append(0)
                continue
            if row['humedad_tarde'] >= 80 and row['nubosidad_tarde'] >= 7:
                guess.append(1)
                continue
            if row['humedad_tarde'] > 75:
                count += 1
            if row['horas_de_sol'] <= 6:
                count += 1
            if row['mm_lluvia_dia'] > 0.7:
                count += 1
            if row['presion_atmosferica_tarde'] <= 1018.5 and row['presion_atmosferica_temprano'] <= 1018.5:
                count += 1
            if row['nubosidad_tarde'] == 8:
                count += 1
            llueve = count/5
            if row['llovieron_hamburguesas_hoy'] == 'no' and (row['horas_de_sol'] >= 9.5 or (row['presion_atmosferica_tarde'] >= 1025 and row['presion_atmosferica_temprano'] >= 1025)):
                guess.append(0)
            elif llueve < 0.5:
                guess.append(0)
            else:
                guess.append(1)  
    return(guess)
```

```python
datos_lluvia_sin_nulls = datos_lluvia[datos_lluvia['llovieron_hamburguesas_al_dia_siguiente'].notna()]
target_list = []
for index, row in datos_lluvia_sin_nulls.iterrows():
    target_list.append(1) if (row['llovieron_hamburguesas_al_dia_siguiente'] == "si") else target_list.append(0)
predicted_list = baseline(datos_lluvia)

accuracy_score(predicted_list, target_list)
```
